```python
import os
import threading
import collections
import time
import random

def count_words(chunk, word_count):
    words = chunk.split()
    for word in words:
        word_count[word] += 1

def main():
    chunk_size = 1024 * 1024  # 1 MB
    file_path = "lorem_5gb.txt"

    file_size = os.path.getsize(file_path)

    try:
        file = open(file_path, "r")

        word_counts = []
        for _ in range(file_size // chunk_size + 1):
            chunk = file.read(chunk_size)
            word_count = collections.defaultdict(int)
            word_counts.append(word_count)
            threading.Thread(target=count_words, args=(chunk, word_count)).start()

        threads = threading.enumerate()
        for thread in threads:
            if thread != threading.current_thread():
                thread.join()

        end = time.time()
        duration = end - start
        print("\nTiempo:", duration, "s.")

        with open("resultados.txt", "w") as outdata:
            outdata.write("Tiempo: " + str(duration) + " s.\n")

            global_word_count = collections.defaultdict(int)
            for word_count in word_counts:
                for word, count in word_count.items():
                    global_word_count[word] += count

            x = len(global_word_count)
            print("\nNro. Palabras:", x)
            outdata.write("\nNro. Palabras: " + str(x) + "\n\n")

            for word, count in global_word_count.items():
                print("{" + word + ": " + str(count) + "}")
                outdata.write("{" + word + ": " + str(count) + "}\n")

    except IOError:
        print("Failed to open file")

if __name__ == "__main__":
    start = time.time()
    main()
```
